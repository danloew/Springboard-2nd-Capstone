## Erroneous Values

As is seen below, there are 30 counties where the age is over 100. In the '2nd_Capstone_Cancer_LogExp_Expansion' notebook, it was noticed that there are 30 cases where the age is well over 100, into the 200-600 year old range. So it is obvious that these values are erroneous, and are replaced by the median value of the 'MedianAge' column in the next series of cells.

```python
cancer['age_gt_100'] = cancer['MedianAge'] > 100
```

```python
cancer['age_gt_100'].value_counts()
```

```python
MedianAge_median = cancer['MedianAge'].median()
MedianAge_median
```

```python
true_median_age = {True: MedianAge_median}
```

```python
cancer.loc[cancer['MedianAge'] > 100,'MedianAge']=np.nan
```

```python
cancer.MedianAge = cancer.MedianAge.fillna(cancer.age_gt_100.map(true_median_age))
```

```python
cancer.MedianAge.value_counts()
```

It was also noticed in the '2nd_Capstone_Cancer_LogExp_Expansion' notebook that there are 61 counties where the average household size for occupied housing units (variable name 'AvgHouseholdSize') is less than one, ranging from 0.0221 to 0.0322. This is again erroneous, as the Census data that this is taken from explicitly states that the 'AvgHouseholdSize' variable stores the average household size for OCCUPIED housing units. Therefore, these counties will have their value filled by the median value in the next series of cells.

```python
cancer['household_lt_1'] = cancer['AvgHouseholdSize'] < 1
```

```python
cancer['household_lt_1'].value_counts()
```

```python
AvgHouseholdSize_median = cancer['AvgHouseholdSize'].median()
AvgHouseholdSize_median
```

```python
true_avghouseholdsize = {True: AvgHouseholdSize_median}
```

```python
cancer.loc[cancer['AvgHouseholdSize'] < 1,'AvgHouseholdSize']=np.nan
```

```python
cancer.AvgHouseholdSize = cancer.AvgHouseholdSize.fillna(cancer.household_lt_1.map(true_avghouseholdsize))
```

```python
cancer.AvgHouseholdSize.value_counts()
```

Essentially, 'avgdeathsperyear' stores the same information as the target variable, just non-normalized.

# Distances to Top 10 Oncology Hospitals and Regional Cities as Predictive Features

The following section computes the L1 and L2 distances of each county from the nation's top 10 oncology hospitals, and from the main large regional cities of the U.S. The top 10 oncology hospitals were from the 2019 ratings found at https://health.usnews.com/best-hospitals/rankings/cancer . The latitudes and longitudes of the top 10 oncology hospitals and of the large regional cities of the U.S. were found from Google search.
